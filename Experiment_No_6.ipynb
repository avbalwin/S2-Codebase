# Aim
Evaluate the line integral of the vector field along a curve using Green’s theorem.

**Problem 1 Evaluate the line integral of the vector field $F(x,y)=(x^2-y,x+y^2)$ along the positively oriented unit circle $C$ defined by $x^2+y^2=1$
 using Green’s theorem.**

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variables
x, y = sp.symbols('x y')

# Define the vector field components
P = x**2 - y
Q = x + y**2

# Compute the partial derivatives
dQ_dx = sp.diff(Q, x)
dP_dy = sp.diff(P, y)

# Apply Green's theorem: double integral of (dQ/dx - dP/dy) over the region D
integrand = dQ_dx - dP_dy

# Define the region D: the unit disk x^2 + y^2 <= 1
# We will use polar coordinates to simplify the integral

r, theta = sp.symbols('r theta')
x_polar = r * sp.cos(theta)
y_polar = r * sp.sin(theta)

# Change of variables: integrand in polar coordinates
integrand_polar = integrand.subs({x: x_polar, y: y_polar}) * r

# Define the limits for r and theta
r_limits = (r, 0, 1)
theta_limits = (theta, 0, 2 * sp.pi)

# Compute the double integral in polar coordinates
area_integral = sp.integrate(integrand_polar, r_limits, theta_limits)

# Display the result
area_integral
```

**Problem 2**

**Evaluate the line integral of the vector field $F(x,y)=(y^2−x,x^2+y)$ along the positively oriented circle $C:x^2+y^2=4$ using Green’s Theorem.**

```python
import sympy as sp

# Define variables
x, y = sp.symbols('x y')

# Define vector field
P = y**2 - x
Q = x**2 + y

# Compute partial derivatives
dQ_dx = sp.diff(Q, x)
dP_dy = sp.diff(P, y)

# Green's theorem integrand
integrand = dQ_dx - dP_dy

# Convert to polar coordinates
r, theta = sp.symbols('r theta')
x_polar = r * sp.cos(theta)
y_polar = r * sp.sin(theta)

integrand_polar = integrand.subs({x: x_polar, y: y_polar}) * r

# Limits: circle of radius 2
area_integral = sp.integrate(
    integrand_polar,
    (r, 0, 2),
    (theta, 0, 2*sp.pi)
)

area_integral
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vector field
def P(x, y):
    return y**2 - x

def Q(x, y):
    return x**2 + y

# Grid
x_vals = np.linspace(-2.5, 2.5, 20)
y_vals = np.linspace(-2.5, 2.5, 20)
X, Y = np.meshgrid(x_vals, y_vals)

U = P(X, Y)
V = Q(X, Y)

# Circle of radius 2
theta = np.linspace(0, 2*np.pi, 200)
x_circle = 2 * np.cos(theta)
y_circle = 2 * np.sin(theta)

# Plot
plt.figure(figsize=(8, 8))
plt.quiver(X, Y, U, V, color='blue', alpha=0.5)

plt.plot(x_circle, y_circle, color='red', linewidth=2, label='x² + y² = 4')
plt.fill(x_circle, y_circle, 'red', alpha=0.2)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field and Circular Region')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.axis('equal')
plt.show()
```
