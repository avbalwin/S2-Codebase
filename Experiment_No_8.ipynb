##Experiment No.8

Solve the homogeneous linear differential equation and visualize

**Case:** Solve the differential equation $y''+4y'+4y=0,y(0)=1,y'(0)=-2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#Define the symbols
t=sp.symbols('t')
y=sp.Function('y')(t)
#Define the differential equation
diff_eqn=sp.Eq(y.diff(t,t)+4*y.diff(t)+4*y,0)
#Solve the differential equation
sol=sp.dsolve(diff_eqn,y)
print("General  solution :",sol)
```

```python
#Apply initial condition
c1,c2=sp.symbols('C1,C2')
y_0=sol.rhs.subs({t:0})
eqn1=sp.Eq(y_0,1)
y_prime=sol.rhs.diff(t)
y_prime0=y_prime.subs({t:0})
eqn2=sp.Eq(y_prime0,-2)
solution=sp.solve((eqn1,eqn2),(c1,c2))
particular_sol=sol.rhs.subs(solution)
particular_solution=sp.Eq(y,particular_sol)
particular_solution
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Convert sympy solution to a numerical function
f=sp.lambdify(t,particular_sol,'numpy')
#Time range for plotting
t_vals=np.linspace(0,10,400)
#Calculate the numerical solutions
y_vals=f(t_vals)
#Plot the solution
plt.plot(t_vals,y_vals,label='y(t)')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution of y\'\'+4y\'+4y=0')
plt.legend()
plt.grid(True)
plt.show()
```

$$m\dfrac{d^2 x}{dt^2}+c\dfrac{dx}{dt}+kx=0$$

$m=1,c=0.5,k=2$

$x(0)=1$ and $x'(0)=0$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#Define the symbols
t=sp.symbols('t')
x=sp.Function('x')(t)
#Define the differential equation
diff_eqn=sp.Eq(x.diff(t,t)+0.5*x.diff(t)+2*x,0)
#Solve the differential equation
sol=sp.dsolve(diff_eqn,x)
print("General  solution :",sol)
```

```python
#Apply initial condition
c1,c2=sp.symbols('C1,C2')
# Ensure t and x are defined in this cell's scope
t=sp.symbols('t')
x=sp.Function('x')(t)
x_0=sol.rhs.subs({t:0})
eqn1=sp.Eq(x_0,1)
x_prime=sol.rhs.diff(t)
x_prime0=x_prime.subs({t:0})
eqn2=sp.Eq(x_prime0,0)
solution=sp.solve((eqn1,eqn2),(c1,c2))
particular_sol=sol.rhs.subs(solution)
particular_solution=sp.Eq(x,particular_sol)
particular_solution
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Convert sympy solution to a numerical function
f=sp.lambdify(t,particular_sol,'numpy')
#Time range for plotting
t_vals=np.linspace(0,10,400)
#Calculate the numerical solutions
y_vals=f(t_vals)
#Plot the solution
plt.plot(t_vals,y_vals,label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Solution of x\'\'+0.5y\'+2y=0')
plt.legend()
plt.grid(True)
plt.show()
```
