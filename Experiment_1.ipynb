#### 2. An electronic temperature sensor outputs a voltage depending on the operating condition of a device. Voltage Classification

• Voltage < 2 V is Low Voltage

• Voltage between 2 V and 5 V is Safe Voltage

• Voltage > 5 V is High Voltage

• Given Sensor output voltage = 3.6 V

### Task
• Store the voltage value in a variable.

• Use conditional statements to classify the voltage level.

• Display the corresponding message.

```python
voltage = 3.6
if voltage < 2:
  print("Low Voltage")
elif voltage <= 5:
  print("Safe Voltage")
else:
  print("High Voltage")
```

#### 3. An EC engineer is analyzing the frequency components present in a signal.Initially, only a few frequency components are detected.

Given

• Initial frequencies (in Hz): [100, 200, 300]

• A new frequency component 400 Hz is detected

• 200 Hz is identified as noise and must be removed.

• A missing low-frequency component 50 Hz must be added at the beginning.

### Task
• Create a list of initial frequencies.

• Add the new frequency at the end of the list.

• Remove the noise frequency.

• Insert the missing frequency at the first position.

• Display the final list of frequencies.

```python
frequencies = [100, 200, 300]
print("Initial Frequencies:", frequencies)
frequencies.append(400)
print("After adding new frequency:", frequencies)
frequencies.remove(200)
print("After removing noise:", frequencies)
frequencies.insert(0, 50)
print("Final Frequency List:", frequencies)
```

#### 4. In circuit analysis, electrical power consumed by a device is calculated using the formula P=V * I where
• V is voltage in volts

• I is current in amperes

Given

• Voltage = 12 V

• Current = 2 A

#### Task

• Define a function to calculate electrical power.

• Pass voltage and current as parameters.

• Return the calculated power.

• Display the power consumed.

```python
def calculate_power(voltage, current):
  return voltage * current
power = calculate_power(12, 2)
print("Power consumed:", power, "Watts")
```
