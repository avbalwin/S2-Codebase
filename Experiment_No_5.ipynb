# Experiment No : 5

Calculate and visualize line integrals.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Define the parameters
t=sp.symbols('t')
#Define the vector field F(x,y,z)=(P,Q,R)
x,y,z=sp.symbols('x y z')
P=x*y
Q=y*z
R=x*z

#Define the parameterized curve r(t)=(x(t),y(t),z(t))
x_t=sp.sin(t)
y_t=sp.cos(t)
z_t=t

#Define the vector field as a function of t along
F = sp.Matrix([P,Q,R])
r_t=sp.Matrix([x_t,y_t,z_t])
F_r_t=F.subs({x:x_t,y:y_t,z:z_t})

#Compute the derivative of r(t) with respect to t
dr_t = sp.diff(r_t,t)

#Compute the dot product of F(r(t)) .dr(t)

integrand = F_r_t.dot(dr_t)

#Define the limits of the parameter t
a =  0 #Start of the parameter range
b = sp.pi #End of the parameter range

#Compute the line integral
line_integral = sp.integrate(integrand,(t,a,b))

# display the result
line_integral
```

**Plotting the vector field and the paramtrized curve**



```python
#Compute the vector field at specific points
F_x, F_y, F_z =F
F_func = sp.lambdify((x, y, z), (F_x, F_y, F_z), 'numpy')

# Generate a grid of points
X, Y, Z = np.meshgrid(np.linspace(-1, 1, 10),
                      np.linspace(-1, 1, 10),
                      np.linspace(0, np.pi,10))

#Compute the vector field at the grid point
U,V,W = F_func(X,Y,Z)

#Compute the vector field at the grid points
t_vals = np.linspace(0,np.pi,100)
x_curve = np.sin(t_vals)
y_curve = np.cos(t_vals)
z_curve = t_vals

#Plot the vector field
fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True, color='blue',alpha = 0.5)

#Plot the curve
ax.plot(x_curve, y_curve, z_curve,color = 'red',linewidth = 2,label = 'Curve r(t)')

#Add labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Field and Parametrized Curve')
ax.legend()

plt.show()
```

```python
#Compute the vector field at specific points
F_x, F_y, F_z =F
F_func = sp.lambdify((x, y, z), (F_x, F_y, F_z), 'numpy')

# Generate a grid of points
X, Y, Z = np.meshgrid(np.linspace(-1, 1, 10),
                      np.linspace(-1, 1, 10),
                      np.linspace(0, 1,10))

#Compute the vector field at the grid point
U,V,W = F_func(X,Y,Z)

#Compute the vector field at the grid points
t_vals = np.linspace(0,1,100)
x_curve = t_vals
y_curve = t_vals**2
z_curve = t_vals**3

#Plot the vector field
fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True, color='green',alpha = 0.5)

#Plot the curve
ax.plot(x_curve, y_curve, z_curve,color = 'red',linewidth = 2,label = 'Curve r(t)')

#Add labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Field and Parametrized Curve')
ax.legend()

plt.show()
```
